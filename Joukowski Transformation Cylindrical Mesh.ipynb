```python
import numpy
import math
from matplotlib import pyplot
%matplotlib inline 
```

```python
n_r = 100
n_theta = 145

r_i = 1.15
r_f = 5.
c = 1.

x_c, y_c = c-r_i, 0.

r = numpy.linspace(r_i,r_f,n_r)
theta = numpy.linspace(0.,2.*numpy.pi,n_theta)

size = 8;
```

```python
r_mesh, theta_mesh = numpy.meshgrid(r, theta) 
```

```python
z = r_mesh*numpy.exp(1j*theta_mesh) + (x_c + 1j*y_c)
x = z.real
y = z.imag
```

```python
pyplot.figure(figsize=(8,8)) 
pyplot.axis('equal')
pyplot.grid(True)
pyplot.title('$z$-plane', fontsize=20)
pyplot.scatter(x,y,s=1.5,color='k',marker='o',linewidth=1);
```

```python
def J_T(xjt,yjt):
    

    
    z = xjt+1j*yjt
    
    zeta = z + 1./z
    
    x_zeta = zeta.real
    y_zeta = zeta.imag
    
    return x_zeta,y_zeta
```

```python
x_zeta,y_zeta = J_T(x,y);
```

```python
z_wing = r_i*numpy.exp(1j*theta) + (x_c + 1j*y_c)  


zeta_wing = z_wing + 1./z_wing  
x_wing = zeta_wing.real
y_wing = zeta_wing.imag
```

```python
pyplot.figure(figsize=(8,8))   
pyplot.grid(True)
pyplot.axis('equal')
pyplot.scatter(x_zeta,y_zeta, s=2,color='k',marker='o',linewidth=0);
```

```python
##freestream streamfunction
u_inf = 1.;

psi_free = u_inf*y;

kappa = r_i**2*u_inf*2*numpy.pi
```

```python
def get_stream_function(r_i, xd, yd, x, y):
    """Returns the strem-function generated by a source/sink.
    
    Arguments
    ---------
    strength -- strength of the source/sink.
    xs, ys -- coordinates of the source/sink.
    X, Y -- mesh grid.
    """
    psi = -(r_i**2*u_inf)*(y-yd)/((x-xd)**2+(y-yd)**2)
    
    return psi
```

```python
psi_doublet = get_stream_function(r_i,x_c,y_c,x,y)
```

```python
psi_total = psi_free + psi_doublet
```

```python
pyplot.figure(figsize=(8,8))
pyplot.grid(True)
pyplot.axis('equal')

levels = numpy.linspace(numpy.min(psi_total), numpy.max(psi_total), 50)
pyplot.contour(x,y,psi_total,levels,colors='k',linewidth=1,linestyles='solid')
```

```python
pyplot.figure(figsize=(8,8))
pyplot.grid(True)

levels = numpy.linspace(numpy.min(psi_total), numpy.max(psi_total), 70)
pyplot.contour(x_zeta, y_zeta, psi_total, levels, colors='k',linewidth=1,linestyles='solid')
pyplot.scatter(x_wing,y_wing,s=75, color='g', marker='o', linewidth=0)
```

```python
def get_velocity_doublet(strength, xd, yd, X, Y):
    """Returns the velocity field generated by a doublet.
    
    Arguments
    ---------
    strength -- strength of the doublet.
    xd, yd -- coordinates of the doublet.
    X, Y -- mesh grid.
    """
    u = - strength/(2.*numpy.pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
    v = - strength/(2.*numpy.pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    
    return u, v
```

```python
u_freestream = u_inf * numpy.ones_like(x);
v_freestream = numpy.zeros_like(y);
```

```python
u_doublet, v_doublet = get_velocity_doublet(kappa,x_c,y_c,x,y)
```

```python
u_total = u_freestream + u_doublet;
v_total = v_freestream + v_doublet;
```

```python
# z = x + jy
# z = zeta +1/zeta
```

```python
uz_iv = u_total - 1j*v_total;

dzeta_dz = 1. - (c/z)**2;

w_zeta = uz_iv / dzeta_dz;

u_zeta = w_zeta.real;
v_zeta = -w_zeta.imag;
u_zeta[0,0]=0.
v_zeta[0,0]=0.
```

```python
step = 3;
window = 3;
pyplot.figure(figsize=(8,8))
pyplot.grid(True)
pyplot.axis('equal')
pyplot.xlim(-window, window)      
pyplot.ylim(-window, window)   
pyplot.title('$z$-plane', fontsize=20)
pyplot.quiver(x[::step,::step], y[::step,::step], u_total[::step,::step], v_total[::step,::step], linewidth=0.1, color = 'k', scale = 30)
```

```python
pyplot.figure(figsize=(8,8))
pyplot.grid(True)
pyplot.axis('equal')
pyplot.xlim(-window, window)      
pyplot.ylim(-window, window)   
pyplot.title('$\zeta$-plane', fontsize=20)
pyplot.quiver(x_zeta[::step,::step], y_zeta[::step,::step], u_zeta[::step,::step], \
              v_zeta.imag[::step,::step], linewidth=0.1, color = 'k', scale = 30)
pyplot.scatter(x_wing,y_wing,s=75, color='g', marker='o', linewidth=0)
```

```python
# z′=[z−(xc+iyc)]e**(−i×AoA)

aoa = 20./360.*(2.*numpy.pi);

z_shifted = (z - (x_c + 1j*y_c ) )*numpy.exp(-1j*aoa) 

x_shifted = z_shifted.real
y_shifted = z_shifted.imag
```

```python
psi_doublet_shifted = get_stream_function(r_i,0,0,x_shifted,y_shifted)
psi_freestream_shifted = u_inf*y_shifted

psi_total_shifted = psi_doublet_shifted + psi_freestream_shifted
```
