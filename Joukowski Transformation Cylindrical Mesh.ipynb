```python
import numpy
import math
from matplotlib import pyplot
%matplotlib inline 
```

```python
n_r = 100;
n_theta = 145;
r_i = 1.15;
r_f = 5;

x_c, y_c = -.15,0

r = numpy.linspace(r_i,r_f,n_r)
theta = numpy.linspace(0,3*numpy.pi,n_theta)
```

```python
r_mesh, theta_mesh = numpy.meshgrid(r, theta) 
```

```python
z = r_mesh*numpy.exp(1j*theta_mesh) + x_c + 1j*y_c
x = z.real
y = z.imag
```

```python
pyplot.figure(figsize=(6,6))   
pyplot.grid(True)
pyplot.axis('equal')
pyplot.scatter(x,y, s=1.5, color='k',marker='o',linewidth=0);
   
```

```python
def J_T(x,y):
    
    z = x+1j*y
    
    zeta = z + 1./z
    
    x_zeta = zeta.real
    y_zeta = zeta.imag
    
    return x_zeta,y_zeta
    
```

```python
x_zeta,y_zeta = J_T(x,y);
```

```python
pyplot.figure(figsize=(6,6))   
pyplot.grid(True)
pyplot.axis('equal')
pyplot.scatter(x_zeta,y_zeta, s=1.5,color='k',marker='o',linewidth=0);
```

```python
##freestream streamfunction
u_inf = 1.;

psi_free = u_inf*y;
```

```python
def get_stream_function(r_i, xd, yd, x, y):
    """Returns the strem-function generated by a source/sink.
    
    Arguments
    ---------
    strength -- strength of the source/sink.
    xs, ys -- coordinates of the source/sink.
    X, Y -- mesh grid.
    """
    psi = -(r_i**2*u_inf)*(y-yd)/((x-xd)**2+(y-yd)**2)
    
    return psi
```

```python
psi_doublet = get_stream_function(r_i,x_c,y_c,x,y)
```

```python
psi_total = psi_free + psi_doublet
```

```python
pyplot.figure(figsize=(5,5))
pyplot.grid(True)
pyplot.axis('equal')

levels = numpy.linspace(numpy.min(psi_total), numpy.max(psi_total), 50)
pyplot.contour(x,y,psi_total,levels,colors='k',linewidth=1,linestyles='solid')
```

```python
pyplot.figure(figsize=(5,5))
pyplot.grid(True)

levels = numpy.linspace(numpy.min(psi_total), numpy.max(psi_total), 70)
pyplot.contour(x_zeta, y_zeta, psi_total, levels, colors='k',linewidth=1,linestyles='solid')
```
