```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
N = 360
r = 1.5;
#r2 = 2;

xc = numpy.zeros((1,N-1))
yc = numpy.zeros((1,N-1))
```

```python
for i in range(N-1):
    xc[0,i] = r*numpy.cos(i)
    yc[0,i] = r*numpy.sin(i)
```

```python
pyplot.scatter(xc, yc, color='g', s=1, marker='o');
pyplot.axis('equal');
pyplot.grid(True);
```
