```python
import numpy
import math
from matplotlib import pyplot
%matplotlib inline
```

```python
N = 51;
vortexnumber = 11;
vortexstrength = 5.0;
x_start,x_end = -2.0,2.0;
y_start,y_end = -0.5,0.5;
x = numpy.linspace(x_start,x_end,N);
y = numpy.linspace(y_start,y_end,N);
X,Y = numpy.meshgrid(x,y);
```

```python
y_v = numpy.zeros((vortexnumber,1))
x_v = numpy.linspace(-2.0,2.0,vortexnumber)
```

```python
pyplot.scatter(x_v, y_v, color='b', s=20, marker='o');
pyplot.axis('equal');
pyplot.grid(True);
```

```python
def get_vortex_velocity(vortexstrength, xv, yv,vortexnumber, X, Y):
    
    """Returns the velocity field (u, v) generated by a line of vortices
    
    Arguments
    ---------
    strength: array with the strength of the sources .
    x_s, y_s: arrays with the coordinates of the source.
    X, Y: mesh grid.
    """
    u=numpy.zeros((N, N))
    v=numpy.zeros((N, N))
    
    for i in range(vortexnumber):
        
        u += vortexstrength/(2*numpy.pi)*(Y-yv[i])/((X-xv[i])**2+(Y-yv[i])**2)
        v += -vortexstrength/(2*numpy.pi)*(X-xv[i])/((Y-yv[i])**2+(X-xv[i])**2)
    
    return u, v
```

```python
def get_infinite_vortex_velocity(vortexstrength, xv, yv,vortexnumber, X, Y):
    
    """Returns the velocity field (u, v) generated by a line of vortices
    
    Arguments
    ---------
    strength: array with the strength of the sources .
    x_s, y_s: arrays with the coordinates of the source.
    X, Y: mesh grid.
    """
    u=numpy.zeros((N, N))
    v=numpy.zeros((N, N))
    
    a =  xv[1]-xv[0];
        
    u = vortexstrength/(2.*a)*numpy.sinh(2.*numpy.pi*Y/a)/(numpy.cosh(2.*numpy.pi*Y/a) - numpy.cos(2.*numpy.pi*X/a)) 
    v = -vortexstrength/(2.*a)*numpy.sin(2.*numpy.pi*X/a)/(numpy.cosh(2.*numpy.pi*Y/a) - numpy.cos(2.*numpy.pi*X/a))
    
    return u, v
```

```python
#Function to get the stream function of a source or sink

def get_vortex_sf(vortexstrength, xv, yv,vortexnumber, X, Y):
    
    """Returns the stream-function generated by a set of sources.
    
    Arguments
    ---------
    strength: array with the strength of the sources .
    x_s, y_s: arrays with the coordinates of the source.
    X, Y: mesh grid.
    """
    psi=numpy.zeros((N, N))
    
    for i in range(vortexnumber):
        
         psi += vortexstrength/(4*numpy.pi)*numpy.log((X-xv[i])**2+(Y-yv[i])**2)
    
    return psi
```

```python
u_upper = 10.0;
u_lower = -1.0;

v_freestream = numpy.zeros((N, N), dtype=float);

u_freestream = u_upper * numpy.ones((N, N), dtype=float);

psi_freestream = u_upper * Y
```

```python
# computes the velocity field
u_vortex, v_vortex = get_vortex_velocity(vortexstrength, x_v, y_v,vortexnumber, X, Y)

# computes the stream-function
psi_vortex = get_vortex_sf(vortexstrength, x_v, y_v, vortexnumber,X, Y)
```

```python
#Let's do superposition and plot the result

# superposition of the source on the freestream
u_total = u_freestream + u_vortex
v_total = v_freestream + v_vortex

psi = psi_freestream + psi_vortex
```

```python

#Now plotting the streamlines using streamplot(), 
#scatter for the source and contour for psi_stagnation

pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

#sources - otherwise known as the airfoil shape
pyplot.scatter(x_v, y_v, s=50, color='g', marker='o', linewidth=0)

#streamlines
pyplot.streamplot(X, Y, u_vortex, v_vortex, density=2, linewidth=1, arrowsize=1.1, arrowstyle='-|>',color='#483D8B');
```
