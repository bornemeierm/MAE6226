```python
#Function to get the velocity components of a sink or source

def get_velocity(strength, x_s, y_s,num_sources, X, Y):
    
    """Returns the velocity field (u, v) generated by a set of sources
    
    Arguments
    ---------
    strength: array with the strength of the sources .
    x_s, y_s: arrays with the coordinates of the source.
    X, Y: mesh grid.
    """
    u=numpy.zeros((N, N))
    v=numpy.zeros((N, N))
    
    for i in range(num_sources):
        
        u += strength[i]/(2*numpy.pi)*(X-x_s[i])/((X-x_s[i])**2+(Y-y_s[i])**2)
        v += strength[i]/(2*numpy.pi)*(Y-y_s[i])/((X-x_s[i])**2+(Y-y_s[i])**2)
    
    return u, v
```

```python
#Function to get the stream function of a source or sink

def get_stream_function(strength, x_s, y_s,num_sources, X, Y):
    
    """Returns the stream-function generated by a set of sources.
    
    Arguments
    ---------
    strength: array with the strength of the sources .
    x_s, y_s: arrays with the coordinates of the source.
    X, Y: mesh grid.
    """
    psi=numpy.zeros((N, N))
    
    for i in range(num_sources):
        
        psi += strength[i]/(2*numpy.pi)*numpy.arctan2((Y-y_s[i]), (X-x_s[i]))
    
    return psi
```
